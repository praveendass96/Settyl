```python
import openai
import random

# OpenAI API key
openai.api_key = "Generate key"



# Initial conversation setup
def initiate_conversation():
    conversation_history = [
        {"role": "system", "content": "You are a negotiation bot. Negotiate the price of a product with the user."},
        {"role": "assistant", "content": "Hello! Let's negotiate the price for the product. How much are you willing to offer?"}
    ]
    return conversation_history

# Function to communicate with OpenAI
def negotiate_with_ai(user_input, conversation_history):
    """
    Sends user input and conversation history to the OpenAI model
    and returns the model's response.
    """
    conversation_history.append({"role": "user", "content": user_input})

    # Corrected API call for the model
    response = openai.chat.completions.create(
        model="gpt-4",  # Use GPT-4 or GPT-3.5
        messages=conversation_history
    )
    bot_reply = response.choices[0].message.content
    conversation_history.append({"role": "assistant", "content": bot_reply})

    return bot_reply

# Price negotiation logic
BOT_INITIAL_PRICE = 1000
MINIMUM_ACCEPTABLE_PRICE = 700
DISCOUNT_RANGE = (50, 100)

def evaluate_user_offer(user_price, current_bot_offer):
    """
    Logic to evaluate user offer and provide a counteroffer or accept/reject.
    """
    if user_price >= current_bot_offer:
        return f"Offer accepted at ${user_price}. Thank you!", current_bot_offer
    elif user_price >= MINIMUM_ACCEPTABLE_PRICE:
        discount = random.randint(*DISCOUNT_RANGE)
        new_bot_offer = current_bot_offer - discount
        # Ensure that the new offer doesn't go below the minimum price
        if new_bot_offer < MINIMUM_ACCEPTABLE_PRICE:
            new_bot_offer = MINIMUM_ACCEPTABLE_PRICE
        return f"Sorry, the lowest price we can offer is ${new_bot_offer}.", new_bot_offer
    else:
        return f"Your offer is too low. How about a lower price of ${MINIMUM_ACCEPTABLE_PRICE}?", current_bot_offer

# Main chatbot conversation loop
def chatbot_conversation():
    conversation_history = initiate_conversation()
    bot_offer = BOT_INITIAL_PRICE

    while True:
        user_input = input("Your input: ")  # Simulate user interaction
        
        # Extract any pricing information from the user input
        try:
            split_input = user_input.split()

            # Extract only the integer parts
            user_price = [int(word) for word in split_input if word.isdigit()]

            user_price = int(user_price[0]) if user_price else None
  
            if user_price:
                bot_response, bot_offer = evaluate_user_offer(user_price, bot_offer)
            else:
                bot_response = negotiate_with_ai(user_input, conversation_history)

            print("Bot:", bot_response)

            if "accepted" in bot_response.lower():
                break
             # Optionally add logic to update bot_offer based on the negotiation
            if "lower price of" in bot_response:
                bot_offer = int(bot_response.split()[-1].split()[0].lstrip('$').rstrip('?'))
        except ValueError:
            bot_response = negotiate_with_ai(user_input, conversation_history)
            print(f"Bot: {bot_response}")


```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()

def analyze_sentiment(user_input):
    """
    Analyzes the sentiment of user input and returns a sentiment score.
    """
    sentiment = sia.polarity_scores(user_input)
    return sentiment['compound']  # Compound score for overall sentiment

def chatbot_conversation_with_sentiment():
    conversation_history = initiate_conversation()
    bot_offer = BOT_INITIAL_PRICE

    while True:
        user_input = input("Your input: ")  # Simulate user interaction
        sentiment_score = analyze_sentiment(user_input)

        # Adjust bot's response based on sentiment
        if sentiment_score > 0.5:
            print("Bot: You seem very polite. I'll give you a better deal!")
            bot_offer -=50
        
        # Normal price negotiation logic
        try:
            split_input = user_input.split()

            # Extract only the integer parts
            user_price = [int(word) for word in split_input if word.isdigit()]
            
            user_price = int(user_price[0]) if user_price else None
            
            if user_price:
                bot_response = evaluate_user_offer(user_price, bot_offer)
            else:
                bot_response = negotiate_with_ai(user_input, conversation_history)

            print("Bot:", bot_response)

            if "accepted" in bot_response.lower():
                break

            if "lower price of" in bot_response:
                bot_offer = int(bot_response.split()[-1].split()[0].lstrip('$').rstrip('?'))
        except ValueError:
            bot_response = negotiate_with_ai(user_input, conversation_history)
            print(f"Bot: {bot_response}")





       


```
